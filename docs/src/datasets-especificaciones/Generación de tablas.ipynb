```python
import pandas as pd
```

```python
df = pd.read_csv("nomina-funcionarios.csv")
```

```python
df
```

```python
def _df_cols_to_th(df):
    th_cols = ["        <th>{}</th>".format(col) for col in df.columns]
    return "\n".join(th_cols)    
```

```python
def _df_rows_to_tr(df):

    td_rows = []
    for row in df.iterrows():
        td_row = "\n        ".join([
            "<td>{}</td>".format(element) 
            for element in list(row[1])
        ])
        td_rows.append(td_row)
    
    tr_rows = [
        """
    <tr>
        {}
    </tr>
        """.format(td_row) for td_row in td_rows
    ]

    return "".join(tr_rows)    
```

```python
def df_to_html(df):
    
    html="""
<table>
    <tr>
{columns}
    </tr>
{rows}
</table>
    """.format(
        columns=_df_cols_to_th(df),
        rows=_df_rows_to_tr(df)
    )
    
    return html
```

```python
for clase in df["clase"].unique():
    print()
    print("===== {} =====".format(clase.ljust(10)))
    print(df_to_html(df[df.clase == clase]))
```
